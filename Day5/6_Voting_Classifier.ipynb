```python
import pandas as pd
from  sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor,  plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,  f1_score, r2_score
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.svm import SVC 
from sklearn.ensemble import VotingClassifier
from tqdm import tqdm
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_selector, make_column_transformer
```

```python
df = pd.read_csv('BreastCancer.csv')
```

```python
df
```

```python
X = df.drop(['Class','Code'], axis = 1)
y = df['Class']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,y,test_size=0.3, random_state=24, stratify=y)
```

```python
dtc = DecisionTreeClassifier(random_state=24)
svc_l = SVC(kernel='linear')
svc_r = SVC(kernel='rbf')

voting = VotingClassifier([('TREE', dtc),('SVC_L', svc_l),('SVC_R', svc_r)])
voting.fit(X_train,Y_train)
y_pred = voting.predict(X_test)

print(accuracy_score(Y_test,y_pred))
```

# Using HR data 

```python
df = pd.read_csv('HR_comma_sep.csv')
df
```

```python
X, y = df.drop('left', axis=1), df['left']
X_train, X_test, Y_train, Y_test = train_test_split(X,y,test_size=0.3, random_state=24)
```

```python
ohe = OneHotEncoder(sparse_output=False,drop='first').set_output(transform='pandas')
ct =  make_column_transformer((ohe, make_column_selector(dtype_include=object)),
                             ('passthrough', make_column_selector(dtype_exclude=object)),
                             verbose_feature_names_out=False).set_output(transform='pandas')

X_dumm = ct.fit_transform(X_train)

X_test_dumm = ct.transform(X_test)
dtc = DecisionTreeClassifier(random_state=24)
svc_l = SVC(kernel='linear')
svc_r = SVC(kernel='rbf')

voting = VotingClassifier([('TREE', dtc),('SVC_L', svc_l),('SVC_R', svc_r)])
voting.fit(X_dumm,Y_train)
y_pred = voting.predict(X_test_dumm)



print(accuracy_score(Y_test,y_pred))

```
