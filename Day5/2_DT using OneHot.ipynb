```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,f1_score
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from tqdm import tqdm
from sklearn.compose import make_column_selector, make_column_transformer
import numpy as np
```

```python
df = pd.read_csv('HR_comma_sep.csv')
```

```python
df
```

```python
X , y= df.drop('left', axis = 1) , df['left']
```

```python
X_dumm = pd.get_dummies(X, drop_first=True)
X_dumm
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_dumm,y,test_size=0.3, random_state=24)
```

```python
max_depths = [2,3,4,None]
min_sample_split = [2,10,100,1000]
min_sample_leaf = [1,50,100,200]
scores = []
for i in tqdm(range(len(max_depths))):
    for j in tqdm(range(len(min_sample_split))):
        for k in tqdm(range(len(min_sample_leaf))):
            td = DecisionTreeClassifier(random_state=24,max_depth=max_depths[i],min_samples_split=min_sample_split[j], min_samples_leaf=min_sample_leaf[k])
            td.fit(X_train,Y_train)
            y_pred = td.predict(X_test)
            scores.append([max_depths[i],min_sample_split[j],min_sample_leaf[k],f1_score(Y_test,y_pred, pos_label=1)])
df_res = pd.DataFrame(scores, columns = ['max_depth', 'Split','leaf','F1'])
df_res.sort_values('F1', ascending= False).iloc[0]            
```

# Using Pipeline 

```python
from sklearn.pipeline import Pipeline

ohe = OneHotEncoder(sparse_output=False, drop= 'first').set_output(transform='pandas')
ct = make_column_transformer((ohe, make_column_selector(dtype_include=object)),
                            ("passthrough", make_column_selector(dtype_exclude=object)),
                            verbose_feature_names_out=False).set_output(transform='pandas')

DT= DecisionTreeClassifier(random_state=24,max_depth=None,min_samples_split=2, min_samples_leaf=1)
pipe = Pipeline([('CT', ct), ('Tree', DT)])
pipe.fit(X_dum, y)
y_pred = pipe.predict(X_test_dum)
```
