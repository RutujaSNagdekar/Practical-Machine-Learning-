```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import numpy as np
from sklearn.decomposition import PCA
import os
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split ,StratifiedKFold , GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv('data.csv')
```

```python
df
```

```python
X , y = df.drop('Bankrupt?', axis = 1) , df['Bankrupt?']
```

```python
X_train , X_test , Y_train, Y_test = train_test_split(X,y,test_size=0.3,random_state=24)
```

```python
lr = LogisticRegression()
scl_std , scl_min  = StandardScaler() , MinMaxScaler()
pca = PCA()
pipe = Pipeline([('scaler', None) ,('PCA', pca) , ('lr' , lr)])
```

```python
pipe.get_params()
```

```python
params = {'scaler' : [scl_std , scl_min, None] ,'PCA__n_components' : np.arange(1,100) , 'lr__C' : [0.01,0.1,0.5,2.5]}
```

```python

kfold = StratifiedKFold(n_splits=5, shuffle = True, random_state=24)
gcv = GridSearchCV(pipe, param_grid= params ,cv = kfold, scoring='f1')
gcv.fit(X_train,Y_train)
```

```python
y_pred = gcv.predict(X_test)
```

```python
print(gcv.best_params_ , gcv.best_score_)
```

```python
print(accuracy_score(Y_test , y_pred))
```

```python

#With Pipeline Only 

from sklearn.metrics import f1_score
comps = [30,35,62,72]
l1  = [0.01,0.1,1]
#comps =[32,48,40,54,62,72]
scores = []

for i in comps:
    for j in l1:
        lr = LogisticRegression(l1_ratio=j)
        scl_min  =  MinMaxScaler().set_output(transform='pandas')
        pca = PCA(n_components = i).set_output(transform='pandas')
        pipe = Pipeline([ ('SCL', scl_min) , ('PCA', pca) , ('LR', lr)])
        pipe.fit(X_train , Y_train)
        y_pred = pipe.predict(X_test)
        scores.append([i ,j, f1_score(Y_test , y_pred , pos_label=1)])
```
