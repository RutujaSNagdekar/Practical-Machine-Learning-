```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor,  plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,  f1_score, r2_score
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split
from tqdm import tqdm 
from sklearn.ensemble import BaggingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
kyp = pd.read_csv('Kyphosis.csv')
kyp
```

```python
X, y = kyp.drop('Kyphosis', axis=1) , kyp['Kyphosis'] 
X_train, X_test, Y_train, Y_test = train_test_split(X,y,test_size=0.3, random_state=24, stratify=y)
```

```python
lr  = LogisticRegression(penalty='l2')
bagg = BaggingClassifier(estimator=lr , n_estimators= 20, random_state=24)
bagg.fit(X_train,Y_train)
y_Pred = bagg.predict(X_test)
print(f1_score(Y_test,y_Pred, pos_label='present'))
#random_state=24  as Bootstrap sample change for machine "
```

```python
#With SVM linear 
svm = SVC(kernel= 'linear')
bagg = BaggingClassifier(estimator=svm,n_estimators=20,random_state=24)
bagg.fit(X_train,Y_train)
y_pred = bagg.predict(X_test)
print(f1_score(Y_test,y_pred,pos_label='present'))
```

```python
#With DT 
dtc = DecisionTreeClassifier(random_state=24)
bagg = BaggingClassifier(estimator=dtc,n_estimators=20,random_state=24)
bagg.fit(X_train,Y_train)
y_pred = bagg.predict(X_test)
print(f1_score(Y_test,y_pred,pos_label='present'))
```

# With Sonar Dataset

```python
df = pd.read_csv('Sonar.csv')
```

```python
df
```

```python
X,y = df.drop('Class', axis = 1), df['Class']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,y,test_size=0.3, random_state=24, stratify=y)
```

```python
# SVM linear
svm = SVC(kernel='linear')
bagg = BaggingClassifier(estimator=svm,n_estimators=20,random_state=24)
bagg.fit(X_train,Y_train)
y_pred = bagg.predict(X_test)
print(f1_score(Y_test,y_pred, pos_label='R'))
```

```python
# SVM Radial
svm = SVC(kernel='rbf')
bagg = BaggingClassifier(estimator=svm,n_estimators=20,random_state=24)
bagg.fit(X_train,Y_train)
y_pred = bagg.predict(X_test)
print(f1_score(Y_test,y_pred, pos_label='R'))
```

```python
# Decision Tree
dtc = DecisionTreeClassifier(random_state=24)
bagg = BaggingClassifier(estimator=dtc,n_estimators=20,random_state=24)
bagg.fit(X_train,Y_train)
y_pred = bagg.predict(X_test)
print(f1_score(Y_test,y_pred, pos_label='R'))
```

```python
# With Logistic Regression
lr  = LogisticRegression(penalty='l2')
bagg = BaggingClassifier(estimator=lr , n_estimators= 20, random_state=24)
bagg.fit(X_train,Y_train)
y_Pred = bagg.predict(X_test)
print(f1_score(Y_test,y_Pred, pos_label='R'))
```
