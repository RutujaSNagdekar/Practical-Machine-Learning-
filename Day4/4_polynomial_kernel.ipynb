```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, r2_score, f1_score
from sklearn.svm import SVC
from tqdm import tqdm
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv('Kyphosis.csv')
```

```python
df
```

```python
X = df.drop('Kyphosis',axis = 1)
y = df['Kyphosis']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=24, stratify=y)
```

```python
scaler = MinMaxScaler()
C = np.linspace(0.001, 4, 10)
ds = [2,3,4]

scores = []

for i in tqdm(range(len(C))):
    for j in ds:
        svm = SVC(kernel= 'poly',C=C[i], degree= j)
        pipe =Pipeline([('scl', scaler),('svm', svm)])
        pipe.fit(X_train,y_train)
        y_pred = pipe.predict(X_test)
        scores.append([C[i],j,f1_score(y_test,y_pred,pos_label='present')])
df1 = pd.DataFrame(scores , columns=['C', 'degree', 'F1'])
df1.sort_values('F1',ascending=False ).iloc[0]      
```
