```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
```

```python
df =  pd.read_csv('BreastCancer.csv')
```

```python
df
```

```python
X = df.drop(['Class', 'Code' ], axis=1)
y= df['Class']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, r2_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=24, stratify=y)
```

```python
svm = SVC(kernel='linear')
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test)
print('Accuracy: ', accuracy_score(y_test, y_pred))
```

```python
C = [0.001, 0.5, 1, 3, 5, 10]
scores = []

for i in C:
    svm = SVC(kernel='linear', C=i)
    svm.fit(X_train, y_train)
    y_pred = svm.predict(X_test)
    scores.append(accuracy_score(y_test, y_pred))

i_max = np.argmax(scores)
print('Max score is ', scores[i_max])
print('K value is ', C[i_max])
print(scores)
```

```python
df = pd.read_csv('HR_comma_sep.csv')
```

```python
df
```

```python
X = df.drop('left',axis = 1)
y = df['left']
```

```python
X_dumm = pd.get_dummies(X)
```

```python
X_dumm
```

```python
from tqdm import tqdm


X_train, X_test, y_train, y_test = train_test_split(X_dumm, y, test_size=0.2, random_state=24, stratify=y)
```

```python

# svm = SVC(kernel='linear')
# svm.fit(X_train,y_train)
# y_pred = svm.predict(X_test)
# print(accuracy_score(y_test,y_pred))
```

```python
C = [0.001,1,2]
scores = []


for i in tqdm(range(len(C))):
    svm = SVC(kernel='linear', C=C[i])
    svm.fit(X_train, y_train)
    y_pred = svm.predict(X_test)
    scores.append(accuracy_score(y_test, y_pred))

i_max = np.argmax(scores)
print('Max score is ', scores[i_max])
print('K value is ', C[i_max])
print(scores)
```
