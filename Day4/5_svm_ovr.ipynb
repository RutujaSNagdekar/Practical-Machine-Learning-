```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, r2_score, f1_score
from sklearn.svm import SVC
from tqdm import tqdm
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_csv('Glass.csv')
df
```

```python
X = df.drop('Type', axis=1)
y = df['Type']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=24, stratify=y)
```

```python
scaler = StandardScaler()

Cs = np.linspace(0.001, 4, 10)

DF = ['ovo', 'ovr']
scores = []


for i in tqdm(range(len(Cs))):
    for j in DF:
        svm = SVC(kernel='linear', C = Cs[i], decision_function_shape=j)
        pipe = Pipeline([('Scaler', scaler) , ('SVM',svm )])
        pipe.fit(X_train, y_train)
        y_pred = pipe.predict(X_test)
        scores.append([Cs[i], j, accuracy_score(y_test, y_pred)])

print(scores)

df1 = pd.DataFrame(scores , columns=['C', 'degree', 'F1'])
df1.sort_values('F1',ascending=False ).iloc[0] 
```

# For Image Segmentation Dataset

```python
df = pd.read_csv('Image_Segmentation.csv')
```

```python
df
```

```python
X = df.drop('Class',axis = 1)
y = df['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=24, stratify=y)
```

```python
#Using Gamma(Radial)

scaler = MinMaxScaler()

Cs = np.linspace(0.001, 4, 10)
Gs = np.linspace(0.001, 4, 10)
DF = ['ovo', 'ovr']
scores = []


for c in tqdm(range(len(Cs))):
    for d in DF:
        for g in Gs:
            svm = SVC(kernel='linear',C=Cs[c],decision_function_shape=d, gamma=g)
            pipe = Pipeline([('scale',scaler),('svm', svm)])
            pipe.fit(X_train,y_train)
            y_pred = pipe.predict(X_test)
            scores.append([Cs[c],d, g ,accuracy_score(y_test,y_pred)])

df1 = pd.DataFrame(scores , columns=['C', 'degree', 'Gamma' ,'accuracy'])
df1.sort_values('accuracy',ascending=False ).iloc[0] 
```
