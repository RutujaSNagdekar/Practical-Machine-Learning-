## ForeCasting Milk Dataset - Trailing Rolling Mean

```python
df = pd.read_csv('monthly-milk-production-pounds-p.csv', index_col=0)
y = df['Milk']
```

```python
y_train = y.iloc[:-12]
y_test = y.iloc[-12:]
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
plt.plot(y_train, c='green')
plt.plot(y_test, c='orange')
```

```python
from numpy import sqrt
span = 3

fcast = y_train.rolling(span).mean()
MA = fcast.iloc[-1]
MA_series = pd.Series(MA.repeat(len(y_test)))
MA_fcast =pd.concat([fcast,MA_series],ignore_index=True)
plt.plot(y_train,label = 'Train')
plt.plot(y_test,label = 'test')
plt.plot(MA_fcast,label='Window Average forecast')
rmse = sqrt(mse(y_test, MA_series))
plt.text(100,600,'RMSE = ' + str(np.round(rmse,2)))
plt.legend(loc='best')
```

# Centered Rolling Mean

```python
from numpy import sqrt
span = 3

fcast = y.rolling(span, center=True).mean()
plt.plot(y,label = 'Train')
plt.plot(fcast,label='Window Average forecast')
plt.legend(loc='best')
```
